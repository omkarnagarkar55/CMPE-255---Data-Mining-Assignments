# Buidl Model Using Pycaret

```python
from pycaret.anomaly import *
s = setup(df, session_id = 123)
```

```python
models()
```

```python
# train cluster model
cluster = create_model('cluster')
cluster
```

```python
cluster_anomalies = assign_model(cluster)
cluster_anomalies
```

# Analyse The Model

```python
evaluate_model(cluster)
```
