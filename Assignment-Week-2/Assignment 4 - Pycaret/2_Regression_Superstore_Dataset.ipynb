```python
# Load dataset from google drive
import pandas as pd
url = "/content/drive/MyDrive/DM-Assignment-Dataset/Superstore.csv"
df = pd.read_csv(url,encoding='windows-1252',error_bad_lines = False)
```

```python
df.columns
```

# Model Building



```python
from pycaret.regression import *
reg_setup = setup(df, target='Profit')
```

```python
# compare models and return the best model
best_model = compare_models()
```

```python
best_model
```

# Analysis

```python
from pycaret.regression import evaluate_model
evaluate_model(best_model)
```

# Test Model

```python
# Prediction on Test Set
prediction = predict_model(best_model)
```

```python
prediction.head()
```

# Save The Model

```python
from pycaret.regression import finalize_model, save_model
final_et = finalize_model(best_model)
save_model(final_et, 'final_et_model')
```

# Load The Saved Model

```python
# load pipeline
from pycaret.regression import load_model
loaded_best_pipeline = load_model('final_et_model')
loaded_best_pipeline
```
