# Model Build Using Pycaret

```python
from pycaret.classification import *
class_setup = setup(df, target='genre')
```

```python
# Comparing different models and returning the best model
best_model = compare_models()
```

# Testing The Model

```python
# Prediction on Test Set
prediction = predict_model(best_model)
```

```python
prediction.head()
```

# Analyse The Model

```python
evaluate_model(best_model)
```

# Save The Model

```python
final_et = finalize_model(best_model)
save_model(final_et, 'final_mc_model')
```
